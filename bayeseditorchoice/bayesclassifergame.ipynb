```python
'''This is a bayes classifier based on classification. Its uni-variate with single attribute to predict 
that will editor choose this came or not based on its user rating. You just have to enter the user 
rating and boom it will show the results based on conditional probability. Data has been taken from 
kaggle of 20 years of gaming world  '''


"""There is no learning rate cause i have used MLE to estimate my parameter that's why there was no use of 
gradient descent to learn and get paramater as it would also give the same output"""


import pandas as pd  #for importing csv file
import numpy as np   #for numeric data   
import requests      #for url
import io   
from numba import vectorize


url="https://raw.githubusercontent.com/kartikpaliwal/bayes-classifer/master/ignew.csv"  #linking the address of csv file
s=requests.get(url).content
dataAddress=pd.read_csv(io.StringIO(s.decode('utf-8')))

TrainingData = dataAddress[0:int(0.75*len(dataAddress))]   #I divided data in two parts
TestingData = dataAddress[(int(0.75*len(dataAddress))):]     


score = list(TrainingData.score)                           #Importing a row of csv file
editorchoice = list(TrainingData.editors_choice)           #Importing a row of csv file


#Here I divided training data in 2 parts i.e. according to 0 or 1

DataDividescoreOne = []    
DataDividescoreZero = []


def EditorScore():
    for i in range(len(score)):
        if editorchoice[i] == 1:
            DataDividescoreOne.append(score[i])
        else:
            DataDividescoreZero.append(score[i])

EditorScore()




#This is maximum likelihood estimation i got this formula by taking derivatives and other stuff. 

def MLE(n):     
    scoreSquare = 0 
    for i in range(len(n)):
        scoreSquare +=(n[i]**2)
    return((1/(2*len(n))*scoreSquare))   


MLE1 = MLE(DataDividescoreOne)   #this is for data of having editorscore 1 
MLE0 = MLE(DataDividescoreZero)  #this is for data of having editorscore 0




#This is the probability distribution function of Rayleigh Distribution.

#As I plotted the frequency plot of the data and by visualization i got that it was like rayleigh.  

def PdfRayleigh(x, MLE10):     
    return ((x/((MLE10)))*(np.e)**(-(np.square(x))/(2*((MLE10)))))





#Testing work starts here!


score_t = list(TestingData["score"])  #This is one row of testing data
editorchoice_t = list(TestingData.editors_choice)   #This is one row of testing data


DataDividescoreOne_t = []       #Here I divided testing data in 2 parts i.e. according to 0 or 1
DataDividescoreZero_t = []


def EditorScore():
    for i in range(len(score_t)):
        if editorchoice_t[i] == 1:
            DataDividescoreOne_t.append(score_t[i])
        else:
            DataDividescoreZero_t.append(score_t[i])

EditorScore()




#This is the process to check accuracy 


ana = []
ananot = []

for r in range (0,len(TestingData)):
    Userscore = (score_t[r])
    a1 = PdfRayleigh(Userscore, MLE1)*(len(DataDividescoreOne_t)/len(TestingData))
    b1 = a1 + PdfRayleigh(Userscore, MLE0)*(len(DataDividescoreZero_t)/len(TestingData))
    n1 = a1/b1

for r in range (0,len(TestingData)):
    Userscore = (score_t[r])
    a2 = PdfRayleigh(Userscore, MLE0)*(len(DataDividescoreZero_t)/len(TestingData))
    b2 = a2 + PdfRayleigh(Userscore, MLE1)*(len(DataDividescoreOne_t)/len(TestingData))
    n2 = a2/b2


#print(n1)
#print(n2)
    
    
    if n1 >= n2:
        Editorchoice = 1
    else :
        Editorchoice = 0


    if Editorchoice == editorchoice_t[r]:
        ana.append(1) 
    else:
        ananot.append(0)   



        
#this is the formulation stuff for calculating the accuracy         

accuracy = ((len(ana))/len(score_t)*100)



#just for viewing

print(len(ana))
print(len(ananot))
print(n1)
print(n2)
```
